```python
import pandas as pd
import csv
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style = "darkgrid")
```

```python
test_set = pd.read_csv("set_for_seaborn2.csv")
```

```python
test_set["Date"] = pd.to_datetime(test_set["Date"], format = "%m/%d/%y")
test_set["New date"] = test_set["Date"].dt.strftime("%b %d")
```

```python
test_set1 = test_set.copy()
test_set1 = test_set1.drop_duplicates(subset = "New date")
```

```python
fig, ax = plt.subplots()
sns.lineplot(test_set.Date, test_set.Amount, hue = test_set.Type)
fig.set_figheight(10)
fig.set_figwidth(15)
ax.set_xticks(["1/5/20", "1/12/20", "1/19/20", "1/26/20", "2/2/20", "2/9/20", "2/16/20", "2/23/20", "3/1/20", "3/8/20"])
ax.set_xticklabels(test_set1["New date"][::7], rotation = 45)
#plt.yscale("log")
plt.savefig('Mediaposts_deaths_time.png')
plt.show()
```
