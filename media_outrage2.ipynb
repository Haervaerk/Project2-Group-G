```python
import pandas as pd
import csv
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style = "darkgrid")
```

```python
test_set = pd.read_csv("set_for_seaborn2.csv")
```

```python
test_set["Date"] = pd.to_datetime(test_set["Date"], format = "%m/%d/%y")
```

```python
test_set
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(test_set.Date, test_set.Amount, hue = test_set.Type).set_title('Media posts and deaths over time')
plt.xticks(["1/5/20", "1/12/20", "1/19/20", "1/26/20", "2/2/20", "2/9/20", "2/16/20", "2/23/20", "3/1/20", "3/8/20"], rotation = 45);
plt.yscale("log")
plt.savefig('Mediaposts_deaths_time.png')
plt.show()
```
