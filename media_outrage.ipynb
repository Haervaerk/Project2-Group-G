```python
import pandas as pd
import csv
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style = "darkgrid")
```

```python
test_set = pd.read_csv("set_for_seaborn1.csv")
```

```python
test_set["Date"] = pd.to_datetime(test_set["Date"], format = "%m/%d/%y")
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(test_set.Date, test_set.Amount, hue = test_set.Type)
plt.xticks(test_set["Date"], rotation = 45)
plt.yscale("log")
```
