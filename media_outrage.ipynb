```python
import pandas as pd
import csv
```

```python
deaths=pd.read_csv('deaths.tsv', sep='\t')
```

```python
len(deaths.columns.values[4:])
```

```python
deaths["total"] = ""
for i, col in enumerate(deaths.columns.values[4:57]):
    deaths["total"][i] = sum(deaths[col])
```

```python
deaths["dates"] = ""
for i in range(len(deaths.columns.values[4:-1])):
    deaths["dates"][i] = deaths.columns.values[4:-1][i]
```

```python
total_death = deaths[["total", "dates"]]
```

```python
total_death = total_death[total_death["total"] != ""]
```

```python
media = pd.read_csv('media_numbers.tsv', sep='\t')
```

```python
total_death["media"] = media["News Articles Published"]
```
